```python
def perceptron(X,y):
    X = np.insert(X,0,1,axis = 1) # insert(array_data, start index, values to be added, axis(0 = rows , 1= columns))
    weights = np.ones(X.shape[1]) # it would create array of 1's depending the features so as given because of
                                  # as given because of insert there are 3 features so there would be o/p [1,1,1] for X.shape[1]
    lr =  0.1

    for i in range (1000):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights))
        weights = weights+lr*(y[j]-y_hat)*X[j]

    return weights[0],weights[1:]
    
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
intercept_,coef_ = perceptron(X,y)
```

```python
print(coef_)
print(intercept_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
